## Análisis exploratorio inicial

```python
import missingno as msno

def diagnostic_summary(df):
    print(f"Dimensiones: {df.shape}")
    summary = pd.DataFrame({
        'dtype': df.dtypes,
        'nulls': df.isnull().sum(),
        'null_pct': (df.isnull().sum() / len(df)) * 100,
        'unique': df.nunique()
    })
    # Visualización de la matriz de nulidad para detectar patrones de correlación en faltantes
    msno.matrix(df)
    plt.show()
    return summary

diagnostic_summary(df)
```
