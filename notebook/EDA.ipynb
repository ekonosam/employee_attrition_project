```python
def assess_stagnation_risk(row):
    """
    Classifies an employee's stagnation risk based on the proportion of time 
    spent in their current role relative to their total tenure, adjusted 
    for their Job Level.
    
    Parameters:
    row (pd.Series): A row from the employee dataframe containing:
                     - 'RoleToCompanyRatio'
                     - 'YearsAtCompany'
                     - 'JobLevel'
                     
    Returns:
    str: A categorical label indicating the risk level.
    """
    
    # 1. Extract variables for readability
    role_tenure_ratio = row['RoleToCompanyRatio']
    total_tenure = row['YearsAtCompany']
    job_level = row['JobLevel']

    # 2. VALIDITY FILTER: Insufficient History
    # Employees with less than 3 years of tenure haven't been around long 
    # enough to establish a meaningful trend of stagnation. 
    # A high ratio here is mathematically normal (e.g., 1 year tenure / 1 year in role).
    if total_tenure < 3:
        return "0. Insufficient History (Safe)"

    # 3. DYNAMIC THRESHOLD EVALUATION
    # We apply different strictness levels depending on seniority.

    # --- JUNIORS (Job Levels 1 & 2) ---
    # Expectation: High rotation and rapid growth.
    # Risk: If they spend most of their short tenure in the same role, they are likely to leave.
    if job_level <= 2:
        # If >80% of their time at the company has been in the same entry-level role.
        if role_tenure_ratio > 0.8: 
            return "CRITICAL: Junior Stagnated (Ratio > 0.8)"
        # If >50% of their time is in the same role.
        if role_tenure_ratio > 0.5: 
            return "WARNING: Junior Slow Progression (Ratio > 0.5)"
    
    # --- SENIORS & LEADERS (Job Levels 3, 4, 5) ---
    # Expectation: Slower movement, but "Entrenchment" is a blocker for the organization.
    # Risk: Because their tenure (denominator) is high, a lower ratio is still significant.
    # A Director with 0.4 ratio (e.g., 8 years in role / 20 years tenure) is heavily entrenched.
    elif job_level >= 3:
        if role_tenure_ratio > 0.5: 
            return "CRITICAL: Senior Entrenched (Ratio > 0.5)"
        if role_tenure_ratio > 0.25: 
            return "WARNING: Senior Momentum Loss (Ratio > 0.25)"

    # Default Case
    return "HEALTHY: Normal Mobility"

# --- APPLICATION ---

# Apply the function row-wise
df['StagnationRiskCategory'] = df.apply(assess_stagnation_risk, axis=1)

# View the resulting contingency table
risk_distribution = pd.crosstab(
    index=df['StagnationRiskCategory'], 
    columns=df['JobLevel']
)

print(risk_distribution)
```

- An analysis with survival analysis with Kaplan-Meirs estimator could be more rigorous and complete and doesn't have the limitations of this table

```python
def calculate_survival_by_group(df, duration_col, is_active_col, group_col):
    """
    Calculates Kaplan-Meier Survival Curves for multiple groups (e.g., JobLevels)
    and handles censoring logic automatically.

    Parameters:
    - df: DataFrame containing employee data.
    - duration_col: The time variable (e.g., 'YearsInCurrentRole').
    - is_active_col: Boolean/String column indicating if employee is still active 
                     (True/'Active' = Censored, False/'Terminated' = Event).
    - group_col: The variable to segment by (e.g., 'JobLevel').

    Returns:
    - full_curves: DataFrame with survival probabilities for every time t per group.
    - median_thresholds: DataFrame with the exact 'Stagnation Point' (median) for each group.
    """

    # --- 1. CENSORING LOGIC ---
    # In Survival Analysis:
    # Event = 1 (The thing happened: Promotion/Exit)
    # Censored = 0 (We stopped watching/They are still there: Active)
    
    # We create a temporary copy to avoid modifying the original
    data = df.copy()
    
    # If the column is boolean (True = Active), Event is the opposite.
    # If it's string ('Yes'/'No'), you might need to adjust this line.
    # Here assuming: True/1 = Still active (Censored), False/0 = Left company (Event)
    data['event_occurred'] = np.where(data[is_active_col] == True, 0, 1)

    # --- 2. THE CALCULATION ENGINE (Applied per Group) ---
    def _compute_km_single_group(subset):
        # Sort by duration is crucial for cumulative math
        # Group by duration to handle ties (multiple people leaving at year 3)
        table = subset.groupby(duration_col)['event_occurred'].agg(['sum', 'count'])
        table.columns = ['events', 'total_observed']
        
        # Calculate 'At Risk' (n_i)
        # Everyone who has duration >= t is at risk at time t
        table['at_risk'] = table['total_observed'].iloc[::-1].cumsum().iloc[::-1]
        
        # Hazard Rate (d_i / n_i)
        # Probability of leaving exactly at time t
        table['hazard'] = table['events'] / table['at_risk']
        
        # Survival Probability S(t) = Cumulative Product of (1 - Hazard)
        # Probability of surviving PAST time t
        table['survival_prob'] = (1 - table['hazard']).cumprod()
        
        return table[['at_risk', 'events', 'survival_prob']]

    # Apply the engine to each JobLevel independently
    full_curves = data.groupby(group_col).apply(_compute_km_single_group)

    # --- 3. EXTRACTING THE "EXACT" STAGNATION THRESHOLDS ---
    # We find the first time 't' where survival probability drops below 0.5 (Median)
    
    thresholds = []
    for group in data[group_col].unique():
        try:
            # Get curve for this group
            curve = full_curves.loc[group]
            
            # Find median time (survival <= 0.5)
            # We use idxmin to find the first index where condition is true, or max time if never reached
            below_median = curve[curve['survival_prob'] <= 0.5]
            
            if not below_median.empty:
                median_time = below_median.index[0]
                status = "Calculated"
            else:
                median_time = curve.index.max()
                status = "Not reached (High Retention)"
                
            thresholds.append({
                group_col: group, 
                'Median_Stagnation_Years': median_time,
                'Status': status
            })
        except KeyError:
            continue
            
    median_thresholds = pd.DataFrame(thresholds).sort_values(group_col)

    return full_curves, median_thresholds

#Create  IsActive column
df['IsActive'] = df['Attrition'] == 'Yes'

# Run the Function
curves, stagnation_report = calculate_survival_by_group(
    df, 
    duration_col='YearsInCurrentRole', 
    is_active_col='IsActive', 
    group_col='JobLevel'
)

# 3. View the "Exact" Thresholds for your strategy
print("--- STAGNATION THRESHOLDS (MEDIAN YEARS) ---")
print(stagnation_report)

# 4. Visualization
plt.figure(figsize=(10, 6))
# We need to iterate through the MultiIndex to plot
for level in df['JobLevel'].unique():
    try:
        subset = curves.loc[level]
        plt.step(subset.index, subset['survival_prob'], where='post', label=f'Job Level {level}')
    except KeyError:
        pass

plt.axhline(0.5, color='red', linestyle='--', label='Median (Stagnation Point)')
plt.title('Survival Curves by Job Level')
plt.ylabel('Probability of Remaining in Role')
plt.xlabel('Years in Role')
plt.legend()
plt.show()
```

### Observations
- The first two job levels their probability drops quickly of remaining in the job while years in role increases. At 2 years in the same role, the job level 1 has alredy 50% of survival rate. This means that it is risky to mantain employees at 2.5 Years or more in that job level. 
- This model gives a measure for when is ideal to raise the job level of an employee if you want to retain it.

```python
# Asumimos que 'matriz_corr' es el resultado numérico de dython que sacaste antes
# Filtramos relaciones fuertes (ej. > 0.6) y que no sean la misma variable (!= 1.0)
strong_pairs = matriz_corr[
    (matriz_corr > 0.4) & (matriz_corr < 1.0)
].stack().index.tolist()

# Eliminamos duplicados (A-B es lo mismo que B-A)
strong_pairs = set(tuple(sorted(p)) for p in strong_pairs)

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

def visualize_relationships_smart(df, pairs_list):
    """
    Iterates through a list of variable pairs [(var1, var2), ...] and automatically 
    generates the most appropriate plot based on data types:
    
    - Numeric vs Numeric: Scatterplot
    - Categorical vs Numeric: Boxplot
    - Categorical vs Categorical: 100% Stacked Bar Chart
    """
    
    for (var1, var2) in pairs_list:
        plt.figure(figsize=(10, 6))
        
        # 1. Detect Data Types
        # We check if columns are Object (strings) or Categorical
        # Note: If your JobLevel is integer (1-5) but acts as a category, cast it to string beforehand.
        is_cat1 = pd.api.types.is_object_dtype(df[var1]) or pd.api.types.is_categorical_dtype(df[var1])
        is_cat2 = pd.api.types.is_object_dtype(df[var2]) or pd.api.types.is_categorical_dtype(df[var2])

        # --- CASE 1: CATEGORICAL vs CATEGORICAL (100% Stacked Bar) ---
        if is_cat1 and is_cat2:
            # A. Create a Contingency Table normalized by index (rows sum to 1 or 100%)
            # This allows us to compare proportions rather than absolute counts.
            normalized_crosstab = pd.crosstab(df[var1], df[var2], normalize='index')
            
            # B. Plot using Pandas built-in plotting (best for stacked bars)
            ax = normalized_crosstab.plot(kind='bar', stacked=True, colormap='viridis', figsize=(10, 6))
            
            plt.title(f"Categorical Relationship: {var1} vs {var2} (Relative Distribution)")
            plt.ylabel("Proportion (0 to 1)")
            plt.xlabel(var1)
            
            # Move legend outside to prevent blocking data
            plt.legend(title=var2, bbox_to_anchor=(1.05, 1), loc='upper left')
            
            # (Optional) Add percentage labels inside the bars
            for n, x in enumerate([*normalized_crosstab.index.values]):
                for (proportion, y_loc) in zip(normalized_crosstab.loc[x], normalized_crosstab.loc[x].cumsum()):                
                    # Only label if segment is large enough (>5%) to be readable
                    if proportion > 0.05:
                        plt.text(x=n,
                                 y=(y_loc - proportion) + (proportion / 2),
                                 s=f'{np.round(proportion * 100, 1)}%', 
                                 color="white", fontsize=9, fontweight="bold", ha="center")

        # --- CASE 2: MIXED (One Categorical, One Numeric) ---
        elif is_cat1 or is_cat2:
            # Ensure Categorical is on X-axis and Numeric on Y-axis for Boxplots
            cat_var = var1 if is_cat1 else var2
            num_var = var2 if is_cat1 else var1
            
            sns.boxplot(x=cat_var, y=num_var, data=df, palette="Set2")
            plt.title(f"Distribution: {cat_var} vs {num_var}")
            plt.xticks(rotation=45)

        # --- CASE 3: NUMERIC vs NUMERIC (Scatterplot) ---
        else:
            sns.scatterplot(x=var1, y=var2, data=df, alpha=0.6)
            plt.title(f"Correlation: {var1} vs {var2}")
            
        plt.tight_layout()
        plt.show()
df_cat= df.copy()
df_cat["JobLevel"]= df_cat["JobLevel"].astype(str )
visualize_relationships_smart(df_cat, strong_pairs)
```

### Is there a lack of monthly income raise to employees with high YearsAtCompany?

### **Econometric Justification for Wage Determinants Modeling**

**Objective:**
The primary objective of this analysis is to isolate and quantify the marginal impact of company-specific tenure on employee compensation. While a simple correlation analysis suggests a weak relationship between longevity and salary, such a view is superficial and potentially misleading due to confounding variables. To address this, i have developed a multivariate log-log regression model designed to enforce a *ceteris paribus* condition. This allows us to determine if the organization effectively incentivizes retention by rewarding loyalty, independent of an employee's total market experience or structural position.

**Methodological Approach:**
By controlling for **Human Capital** (e.g., `TotalWorkingYears`, `Education`, `Age`) and **Structural Hierarchies** (e.g., `JobLevel`, `JobRole`), the model disentangles the "Seniority Premium" from the "Promotion Premium."
Crucially, the inclusion of `JobLevel` and `TotalWorkingYears` ensures that the coefficient for `YearsAtCompany` reflects the true monetary value the company assigns to retention itself. If this coefficient is low or statistically insignificant after controls, it signals a risk of **wage compression**, where long-tenured employees are not compensated competitively compared to new hires or the external market, directly increasing the risk of attrition among high-value veteran staff.

**Model Specification:**
The final econometric specification is defined as follows:

$$\begin{aligned}
\ln(\text{MonthlyIncome}) = & \beta_0 + \beta_1 \ln(\text{YearsAtCompany}) + \beta_2 \text{TotalWorkingYears} \\
& + \beta_3 \text{Age} + \beta_4 \text{NumCompaniesWorked} + \beta_5 \text{JobLevel} \\
& + \beta_6 \text{PerformanceRating} + \sum \delta_i \text{CategoricalControls}_i + \epsilon
\end{aligned}$$


```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import statsmodels.stats.api as sms
from statsmodels.stats.diagnostic import linear_reset
from scipy.stats import anderson

# 1. Preprocesamiento: Transformación logarítmica
# Suponiendo que 'df' es tu DataFrame
df['log_MonthlyIncome'] = np.log(df['MonthlyIncome']+1)

# 2. Definición de la fórmula (statsmodels maneja automáticamente las variables categóricas)
formula = 'log_MonthlyIncome ~  C(JobRole) + YearsAtCompany +I(YearsAtCompany**2) + C(Education) + C(EducationField)  +np.log(TotalWorkingYears+0.01)+ np.log(Age+0.01) + NumCompaniesWorked + JobLevel +OverTime + PerformanceRating'

# 3. Ajuste del modelo OLS
model = smf.ols(formula=formula, data=df).fit(cov_type='HC3')  # Usando errores estándar robustos

print(model.summary())

# --- PRUEBAS DE DIAGNÓSTICO ---

# A. Prueba de Ramsey RESET (Especificación de la forma funcional)
reset_test = linear_reset(model, power=2, test_type="fitted")
print(f"\nRamsey RESET Test (p-value): {reset_test.pvalue:.4f}")

# B. Prueba de Breusch-Pagan (Heteroscedasticity)
names = ['Lagrange multiplier statistic', 'p-value', 'f-value', 'f p-value']
bp_test = sms.het_breuschpagan(model.resid, model.model.exog)
print(f"Breusch-Pagan Test (p-value): {bp_test[1]:.4f}")

# C. Prueba de Anderson-Darling (Normalidad de los residuos)
ad_test = anderson(model.resid, dist='norm')
print(f"Anderson-Darling Statistic: {ad_test.statistic:.4f}")
print(f"Critical Values at 5%: {ad_test.critical_values[2]}")
```

### Observations
to understand better what is the relationship with yearsAtCompany and MonthlyIncome i'll make a graph that resumes the non-linear relationship that i've found

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Define coefficients from the Regression Model
intercept = 8.5115
beta_years = 0.0069
beta_years2 = -0.0003

# 2. Create the range for the X-axis (Tenure from 0 to 40 years)
x_years = np.linspace(0, 40, 100)

# 3. Calculate Prediction (Ceteris Paribus)
# We calculate log_income based only on YearsAtCompany and its quadratic term.
# All other variables (JobRole, Education, etc.) are held constant at 0 (reference category).
log_income_predicted = intercept + (beta_years * x_years) + (beta_years2 * x_years**2)

# 4. Inverse Transformation (Exponential) to get Real Currency ($)
y_income_predicted = np.exp(log_income_predicted)

# 5. Calculate the Inflection Point (Peak)
# Derivative: b1 + 2*b2*x = 0  =>  x = -b1 / (2*b2)
peak_year = -beta_years / (2 * beta_years2)
peak_income = np.exp(intercept + (beta_years * peak_year) + (beta_years2 * peak_year**2))

# --- PLOTTING (ENGLISH) ---
plt.figure(figsize=(10, 6))

# Main Curve
plt.plot(x_years, y_income_predicted, color='#2c3e50', linewidth=3, label='Predicted Salary (Ceteris Paribus)')

# Peak Line
plt.axvline(x=peak_year, color='#e74c3c', linestyle='--', label=f'Peak Earning Year ({peak_year:.1f} years)')
plt.scatter(peak_year, peak_income, color='#e74c3c', s=100, zorder=5)

# Aesthetics & Labels
plt.title('Diminishing Returns of Tenure on Salary', fontsize=14, fontweight='bold')
plt.xlabel('Years at Company', fontsize=12)
plt.ylabel('Predicted Monthly Income ($)', fontsize=12)
plt.legend(loc='upper right')
plt.grid(True, alpha=0.3)

# Annotation box
plt.annotate(f'Max Salary: ${peak_income:,.0f}', 
             xy=(peak_year, peak_income), 
             xytext=(peak_year+3, peak_income),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=10)

plt.tight_layout()
plt.show()
```
